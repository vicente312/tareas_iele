Instrucciones: Durante las cuarentenas, una pregunta que se hizo el Ministerio de Transporte fue si las personas usaban más el auto, o caminaban, y si los parques se visitaban más. Apple y Google sacaron los "mobility" reports, que tienen, a nivel de region, la movilidad de la población.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
google= pd.read_csv('Global_Mobility_Report.csv')
apple = pd.read_csv('applemobilitytrends-2021-12-07.csv')
```

```python
google= google[google['country_region_code']=='CL']
```

```python
google['date'].sort_values()
```

```python
park=google[['date','parks_percent_change_from_baseline',]]
park.head()
```

```python
plt.figure(figsize=(8,6))
ax = sns.lineplot(x="date", y="parks_percent_change_from_baseline",markers=True, dashes=True, palette='RdBu_r', data=park)
plt.axvline('2020-03-20',color="red")
plt.axvline('2020-08-01',color="red")
plt.show()
```

Conclusión: ¿Durante la cuarentena fueron más visitados los parques? 
Según las fecha que nosotros definimos como cuarentena (se pueden ver por las líneas rojas) podemos concluir que no, las personas no visitaban más el parque en cuarentena. Ya que, claramente los valores dentro del rango definido son más bajos que en los períodos sin cuarentena.

```python
apple_cl=apple[apple['region']=='Chile']
apple_cl=apple_cl.drop(['geo_type','region','alternative_name','sub-region','country'],axis=1)
```

```python
x=apple_cl.melt(id_vars=['transportation_type'],var_name='Fecha',value_name='valores')
```

```python
plt.figure(figsize=(8,6))
ax = sns.lineplot(x="Fecha", y="valores",hue="transportation_type",markers=True, dashes=True, palette='RdBu_r', data=x)
plt.axvline('2020-03-20')
plt.axvline('2020-08-01')
plt.show()
```

¿La gente caminaba más o usaba más el auto dependiendo de si estaban en cuarentena o no? 
Respuesta: Según la fecha que definimos como cuarentena, representada por las líneas azules, se puede concluir que no. Las personas no caminaron más ni tampoco usaron más el auto durante la cuarentena. En el gráfico se puede ver claramente que son los valores más bajos dentro del tiempo.
